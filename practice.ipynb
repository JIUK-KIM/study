# 파이썬 머신러닝 판다스 데이터분석

## PART 1 판다스 입문

```python
### <1-1> 딕셔너리 -> 시리즈 변환 ###

    # 판다스 불러오기
import pandas as pd

    # 딕셔너리 선언 및 변수 dic_data에 저장
dic_data = {"a" : 1, "b" : 2, "c" : 3}

    # 판다스 함수 Series()로 dictionary를 Series로 변환 및 변수 sr에 저장
sr_dic = pd.Series(dic_data)

    # 타입 출력
print(type(sr_dic))
print('\n')

    # 시리즈 sr 출력
print(sr)
print('\n')


### --------------------------- ###
### 속성 index와 속성 values로 파악 ###

    #
idx = sr.index
val = sr.values

print(idx)
print(val)
print('\n')

### --------------------------- ###
### 인덱싱, 슬라이싱 ###
print('<인덱싱, 슬라이싱>')

print(sr_dic['a'])
print('\n')

print(sr_dic[['a','c']])
print('\n')

print(sr_dic['b':'c'])
print('\n')
```

```python
### <1-2> 리스트 -> 시리즈 변환 ###
    
    #
import pandas as pd

list_data = ['2019-01-02', 3.14, 'ABC', 100, True]

sr_list = pd.Series(list_data)

print(type(sr_list))
print('\n')

print(sr_list)
print('\n')



### 속성 index와 속성 values로 파악 ###

    #
idx = sr_list.index
val = sr_list.values

    #
print(idx)
print(val)
print('\n')

### --------------------------- ###
### 인덱싱, 슬라이싱 ###
print('<인덱싱, 슬라이싱>')

print(sr_list[0])
print('\n')

print(sr_list[[2,3]])
print('\n')

print(sr_list[0:4])
print('\n')
```

#### 딕셔너리와 달리, 리스트는 key값이 없기 때문에 인덱스로 변환될 값이 없다. 따라서, 인덱스를 별도로 정의하지 않는다면, 디폴트로 정수형 위치(integer index)가 인덱스로 정의된다.

```python
### <1-3> 시리즈 원소 선택 ###

import pandas as pd

tup_data = ('영인', '2010-05-01', '여', True)

sr_tup = pd.Series(tup_data, index=['이름','생년월일','성별','학생여부'])

print(sr_tup)
print('\n')


### 속성 index와 속성 values로 파악 ###

    #
idx = sr_tup.index
val = sr_tup.values

    #
print(idx)
print(val)
print('\n')

### --------------------------- ###
### 인덱싱, 슬라이싱 ###
print('<인덱싱, 슬라이싱>')


print(sr_tup['이름'])
print('\n')

print(sr_tup[['생년월일','학생여부']])
print('\n')

    # 슬라이싱에서, 정수형 인덱스는 초산말불 vs. 라벨형 인덱스는 초산말산
print(sr_tup['생년월일':'학생여부'])
print('\n')

print(sr_tup[1:3])
print('\n')
```
